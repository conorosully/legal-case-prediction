```python
#Clean cases
cleaned = {}
count = 0
for ID in idForm.keys():
    
    case = clean.getCase(ID)[ID]
    form = idForm[ID]
    
    try:
        cleaned[ID] = clean.cleanCase(case,form)
    except:
        print(ID)
    
    
    if count%500 == 0: 
        print("Processed: {}/8703".format(count))
    count += 1
```
