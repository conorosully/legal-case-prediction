```python
#Check if cases are the correct form
print("Form: ")

casenames = []
count = 0
for i in range(len(link)):
    
    casename = link.iloc[i]['doc-name']
    
    ID = 'none'
    IDs = att[att['doc-name'] == casename]['id']
    
    formMatch = False
    
    for ID in IDs:
        case = clean.getCase(ID)[ID]
        form = clean.getForm(case,forms)
        try:
            if clean.checkForm(form): 
                formMatch = True
                link.loc[i,'id'] = ID
                break;
        except:
            #print(i, casename)
            x = 1
    if formMatch == False:
        print(count,i,ID,casename)
        casenames.append(casename)
        count +=1
        
        
print('Clean:')
for i in range(554,len(link)):
    ID = link.iloc[i]['id']
    case = clean.getCase(ID)[ID]
    form = clean.getForm(case,forms)
    try: 
        clean.cleanCase(case,form)
    except:
        print(i,ID,link.iloc[i]['doc-name'])
        break;
        

link.head()
```
